```python
import os
import matplotlib.pyplot as plt
from IPython.display import display
from PIL import Image

# --- Configuration ---
#folder_path = '/explore/nobackup/people/sjaddu/test_images/7-30-2025/satvision/'
folder_path = '/explore/nobackup/people/sjaddu/test_images/UNET_PNG_PREDICTIONS/'

# --- Load and Display All PNGs ---
png_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]

for fname in sorted(png_files):
    img_path = os.path.join(folder_path, fname)
    img = Image.open(img_path)
    
    plt.figure(figsize=(20, 20))  # Adjust figure size as needed
    plt.imshow(img)
    plt.axis('off')
    plt.title(fname)
    plt.show()
```
